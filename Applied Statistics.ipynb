**The ice hockey team Klovnit has a squad consisting of three goalkeep-
ers, seven defenders, ten forwards, and four players who can play both as defenders
and as forwards. A lineup consists of one goalkeeper, two defenders and three
forwards. In how many ways can a lineup be selected from Klovnit's squad?**

Let $G$ represent the goalkeepers, $D$ represent the players who can only be defenders, $F$ represent the players who can only be forwards, and $W$ represent the players who can play either defenders or forwards.

For each goalkeeper the following shows the number of ways to form the lineup:



If the defense has 2 Ds, then you have 14 options for the forwards. If the defense has 2 Ws, then you have only 12 options for the forwards. If the defense has 1 D and 1 W, then there are 13 options for the forwards.

For each goal keeper, there are $\left(_2^7\right)\cdot \left(_{\:3}^{14}\right)+\left(_2^4\right)\cdot \left(_{\:3}^{12}\right)+\left(\left(_{\:2}^{11}\right)-\left(_2^7\right)-\left(_2^4\right)\right)\cdot \left(_{\:3}^{13}\right)=16972$ ways of forming the lineup.

Multiply by 3 for each goalkeeper and the total number of ways $=50916$

In how many ways can n people be seated around a circular table, if
two seatings are considered the same if each person has the same two neighbours?**



There are $n!$ permutations of seating people in a line. Divide by n to get the number of ways to seat people in a circle where rotating the seats but keeping the order is considered the same arrangement (A=B). Since the reverse of the circle is the same seating arrangement, divide by 2 to get the number of unique seating placements (A=C).

Total $=\frac{n!}{n}\cdot \frac{1}{2}$

A sports club offers classes in 
floorball, handball, dancing and skiing,
each having 30 participants. 12 people are playing both
floorball and handball, 10 people are both playing some ball sports and dancing,
and 14 people are both skiing and doing some other sport. How many people are
participating in some of the club's activities?**

Let F, H, D, and S represent the set of students in floorball, handball, dance, and skiing respectively.

$n\left(F\right)=30,\:n\left(H\right)=30,\:n\left(D\right)=30,\:n\left(S\right)=30$

$n\left(F\cap H\right)=12$

$n\left(F\cap D\right)\cup n\left(H\cap D\right)=10$

$n\left(S\cap D\right)\cup n\left(S\cap F\right)\cup n\left(S\cap H\right)=14$

$n\left(F\cup H\right)=n\left(F\right)+n\left(H\right)-n\left(F\cap H\right)=30+30-12=48$

$n\left(S'\right)=n\left(S\right)-n\left(S\cap D+n\left(S\cap H\right)+n\left(S\cap F\right)\right)=30-14=16$

$n\left(D'\right)=n\left(D\right)-\left(n\left(F\cap \:D\right)\cup \:n\left(H\cap \:D\right)\right)=30-10=20$

$n\left(S'\right)+n\left(D'\right)+n\left(F\cup H\right)=total\:=\:16+20+48=84$

At a casino there are two kinds of slot machines, which look quite indentical
from the outside. In machines of type A, the winning probability is 5%. In machines of
type B, the winning probability is 7%. We also know that 70% of the machines are of type A,
and 30% are of type B.
(a) A gambler picks a slot machine at random, and plays once. What is the probability that
he wins?
(b) If the gambler wins, what is the probability that the machine is of type B?**

(a)



The probability of choosing A times the probability of winning if A is chosen, plus the probability of choosing B times the probability of winning if B is chosen:

$0.7\cdot\:0.05+0.3\cdot\:0.07=0.056$

The probability of winning is 0.056.

(b) 

$\frac{.07\cdot .3}{.05\cdot \:.7+.07\cdot \:.3}=0.375$

The probability of the machine being type B if the gambler wins is 0.375.
